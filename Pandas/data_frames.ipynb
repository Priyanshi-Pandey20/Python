```python
import numpy as np
import pandas as pd
```

CREATING DATAFRAMES WITH DICTONARY

```python
data = {
    'Name':['John','Anna','Peter','Linda'],
    'Age':[28,34,29,42],
    'City':['New York','Paris','Berlin','London'],
    'Salary':[65000,70000,62000,85000]
}
pd.DataFrame(data)
```

 CREATING DATAFRAMES WITH DATA LIST

```python
data_list = [
    ['John',	28,	'New York',	65000],
    ['Anna',	32	,'Paris',	70000],
    ['Peter',29	,'Berlin',	62000],
    ['Linda',	28,	'London',	85000],

]
df2 =pd.DataFrame(data_list) #default column name 
print(df2)
columns = ["Name","Age","City","Salary"] # by this we can give headers to the list 
df2 = pd.DataFrame(data_list,columns=columns)
print("\n",df2)
```

SELECTION AND INDEXING OF COLUMNS

```python
df2['Name']
```

```python
df2[['Name','Age']]
```

CREATING NEW COLUMN

```python
df2["Designation"] = ["CA","Architech","Doctor","Designer"]
df2
```

REMOVING A COLUMN

```python
df2.drop('Designation',axis=1) #drop  does not remove the data in real from database (axis=0 for row,axis=1 for col)
df2
```

```python
df2.drop('Designation',axis=1,inplace=True) #inplace completelty removes the data from database
df2
```

SELECTING A ROW


```python
df2.loc[0]  #loc means location 
```

```python
df2.loc[[0,1]] #selecting multiple rows 
```

SELECTING SUBSETS OF A ROWS AND COLUMNS

```python
df2.loc[[0,1]][["City","Salary"]] #first we select rows and then columns
df2.loc[[2,3]][["Name","Age"]]
```

CONDITIONAL SELECTION

```python
#I only want to see whose age is above 30

df2[df2["Age"] > 30]
```

```python
#I only want people whose age is above 20 and their city must be paris

df2[(df2["Age"] > 30) & (df2["City"] == "Paris")] 
```
