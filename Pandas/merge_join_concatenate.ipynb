```python
import numpy as np
import pandas as pd
```

MERGING 2 DATAFRAMES

```python
employees = pd.DataFrame({
    'employee_id':[1,2,3,4,5],
    'name':['John','Anna','Peter','Linda','Bob'],
    'department':['HR','IT','Finance','IT','HR']
})

salaries = pd.DataFrame({
    'employee_id':[1,2,3,6,7],
    'salary':[60000,80000,65000,70000,90000],
    'bonus':[5000,10000,7000,8000,12000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries,on='employee_id',how='inner') #merging the datframes on the basis of inner values
```

```python
pd.merge(employees,salaries,on='employee_id',how='outer')  #merging the datframes on the basis of headers
```

```python
pd.merge(employees,salaries,on='employee_id',how='left') #merging on basis of employee
```

```python
pd.merge(employees,salaries,on='employee_id',how='right')#merging on basis of salary
```

CONCATINATION OF 2 DATAFRAMES

```python
df1 = pd.DataFrame({
    'A':['A0','A1','A2'],
    'B':['B0','B1','B2'],
    'C':['C0','C1','C2'],
})

df2 = pd.DataFrame({
    'A':['A3','A4','A5'],
    'B':['B3','B4','B5'],
    'C':['C3','C4','C5'],
})

```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2]) #df1 is added with df2
```

```python
pd.concat([df2,df1])
```

```python
pd.concat([df2,df1],axis=1)#concatenate on basis of row 
```

JOINING 2 DATAFRAMES

```python
df1 = pd.DataFrame({
    'name':['Alice','Bob','Charlie']
},index=[1,2,3])

#Second DataFrames
df2 = pd.DataFrame({
'score':[85,90,75]
},index=[2,3,4])

```

```python
df1
```

```python
df2
```

```python
df1.join(df2,how='outer')
```

```python
df2.join(df1,how='outer')
```

```python
df2.join(df1)
```
