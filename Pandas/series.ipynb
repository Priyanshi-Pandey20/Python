```python
import numpy as np
import pandas as pd
```

HOW TO CREATE SERIES

```python
labels = ['a','b','c']
my_list =[10,20,30]
arr = np.array([10,20,30])
d = {1:10,2:20,3:30}
```

```python
#creating series with the help of list
pd.Series(my_list)

```

```python
pd.Series(my_list,index=labels) #assigning index with the labels 
```

```python
pd.Series(arr)
```

```python
pd.Series(d)
```
