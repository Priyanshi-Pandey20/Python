GROUP BY

```python
data = {
    'Category':['A','B','A','B','A','B','A','B'],
    'Store':['S1','S1','S2','S2','S1','S2','S2','S1'],
    'Sales':[100,200,150,250,120,180,200,300],
    'Ouantity':[10,15,12,18,8,20,15,25],
    'Date':pd.date_range('2023-01-01',periods=8)

}
df = pd.DataFrame(data)
df
```

```python
#Group by category and calculate the sum of sales

cat = df.groupby('Category')['Sales'].sum() #groupby creates the object which we can't see
cat
```

```python
##Group by category and calculate the sum of store

cat = df.groupby('Category')['Store'].sum()
cat
```

```python
#group by category and store
cat = df.groupby(['Category','Store'])['Sales'].sum()
cat
```
