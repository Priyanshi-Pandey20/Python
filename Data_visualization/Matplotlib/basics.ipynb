```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11) #create spaces upto limit 11 
x
y = x**2
y
```

```python
plt.plot(x,y)
plt.title("Hello")
plt.xlabel(" X axis ")
plt.ylabel("Y axis")
```

SUBPLOTS

```python
plt.subplot(2,2,1)
plt.plot(x**10,y)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,4)
plt.plot(x,y) #if you want to plot in any of them right just below it 
```

OBJECT ORIENTED WAY # when we want to place the one graph on other graph

```python
fig = plt.figure(figsize=(6,8),dpi=100) #6:width,8:height #dpi used to make the graph more clearer 
#figsize is used to increase the size                                        
axis1 = fig.add_axes([0.1,0.1,1,1])#( toshift => left,bottom,width,height)
axis1.plot(x,y)
axis2 = fig.add_axes([0.1,0.1,0.3,0.3])
axis2.plot(x,y)
```

TYPES OF PLOT

SCATTER PLOT

```python
plt.scatter(x,y)
```

HISTOGRAM PLOT

```python
from random import sample
data = sample(range(1,1000),100)
plt.hist(data)
```

BOX PLOT

```python
plt.boxplot(x)
```

SAVING PLOTS

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("BasicPlot.png")
```
