DISTRIBUTION PLOTS : It is used to visualize the quantitative data.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df =sns.load_dataset('tips')
df
```

```python
df['size'].unique
```

DISTPLOT

```python
#kde(kernal density estimator : I t tells whether your graph is left skewed or right skewed ot normally distributed)
sns.histplot(df['total_bill'],bins = 20,kde =True) #bins defines how you can divide the data 
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins = 20,kde =True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins = 20,kde =True)
```

JOINTPLOT

```python
sns.jointplot(x = 'total_bill',y='tip',data=df,kind='scatter')
```

```python
sns.jointplot(x = 'total_bill',y='tip',data=df,kind='hex')
```

```python
sns.jointplot(x = 'total_bill',y='tip',data=df,kind='kde')
```

PAIRPLOT

```python
sns.pairplot(df,hue='size',palette='rainbow')
```

RUGPLOT

```python
sns.rugplot(df['total_bill'])
```
